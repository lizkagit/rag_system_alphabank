```python
# 2. –ó–ê–ì–†–£–ó–ö–ê –î–ê–ù–ù–´–•
print("\nüìÅ –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ...")
# –ü—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º, —á—Ç–æ —Ñ–∞–π–ª—ã –≤ —Ç–æ–π –∂–µ –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏
websites_df = pd.read_csv('websites.csv')
questions_df = pd.read_csv('questions.csv')

print(f"üìä –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(websites_df)} –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ –∏ {len(questions_df)} –≤–æ–ø—Ä–æ—Å–æ–≤")
```

```python
# 3. –ü–û–î–ì–û–¢–û–í–ö–ê –¢–ï–ö–°–¢–û–í –î–õ–Ø –í–ï–ö–¢–û–†–ò–ó–ê–¶–ò–ò
print("\nüîß –ü–æ–¥–≥–æ—Ç–∞–≤–ª–∏–≤–∞–µ–º —Ç–µ–∫—Å—Ç—ã...")

# –ü—Ä–æ—Å—Ç–æ–π –ø–æ–¥—Ö–æ–¥: –æ–±—ä–µ–¥–∏–Ω—è–µ–º title –∏ text
documents = []
for idx, row in websites_df.iterrows():
    combined_text = f"{row['title']}. {row['text']}"
    documents.append(combined_text)

questions = questions_df['query'].tolist()
```

```python
# –ü–æ—Å–º–æ—Ç—Ä–∏–º –Ω–∞ –ø—Ä–∏–º–µ—Ä—ã
print(f"üìÑ –ü—Ä–∏–º–µ—Ä –¥–æ–∫—É–º–µ–Ω—Ç–∞: {documents[0][:100]}...")
print(f"‚ùì –ü—Ä–∏–º–µ—Ä –≤–æ–ø—Ä–æ—Å–∞: {questions[0]}")
```
